```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Load your dataset
df_economy = pd.read_csv("D:\\Pakistan economy and GDP Sectors(1).csv")
df_economy.head()

# Describe the dataset
df_economy.describe()

# Check for missing values
print(df_economy.isnull().sum())
```

```python
# Distribution plot for GDP Current Prices


plt.title('GDP Current Prices Distribution Plot')
sns.histplot(df_economy['GDP Current Prices'])
plt.show()

# Relationship between Year and GDP Current Prices
plt.scatter(df_economy['Year'], df_economy['GDP Current Prices'], color='black')
plt.title('GDP Current Prices vs Year')
plt.xlabel('Year')
plt.ylabel('GDP Current Prices')
plt.box(False)
plt.show()
```

```python
# Splitting variables
X = df_economy[['Year']]  # independent variable
y = df_economy[['GDP Current Prices']]  # dependent variable

# Splitting dataset into test/train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

```

```python
# Regressor model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Prediction result
y_pred_test = regressor.predict(X_test)  # predicted value of y_test
y_pred_train = regressor.predict(X_train)  # predicted value of y_train
```

```python
# Prediction on training set
plt.scatter(X_train, y_train, color='lightcoral')
plt.plot(X_train, y_pred_train, color='firebrick')
plt.title('GDP Current Prices vs Year (Training Set)')
plt.xlabel('Year')
plt.ylabel('GDP Current Prices')
plt.legend(['Predicted', 'Actual'], loc='best')
plt.box(False)
plt.show()

# Prediction on test set
plt.scatter(X_test, y_test, color='lightcoral')
plt.plot(X_train, y_pred_train, color='firebrick')
plt.title('GDP Current Prices vs Year (Test Set)')
plt.xlabel('Year')
plt.ylabel('GDP Current Prices')
plt.legend(['Predicted', 'Actual'], loc='best')
plt.box(False)
plt.show()
```

```python
# Regressor coefficients and intercept
print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred_test)
mse = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

print(f'Mean Absolute Error: {mae}')
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```
