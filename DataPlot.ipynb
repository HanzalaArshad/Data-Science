```python

import pandas as pd

import matplotlib.pyplot as plt

# Load the dataset
economy = pd.read_csv("D:\\Pakistan economy and GDP Sectors(1).csv")

# Extracting years and growth rate from the dataset
years = economy['Year']
growth_rate = economy['Total Agriculture']

# Plotting the graph
plt.figure(figsize=(10, 6))  # Adjust figure size if necessary
plt.plot(years, growth_rate, marker='o', linestyle='-')

# Adding labels and title
plt.title('Pakistan Economy Growth Rate Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Agriculture')

# Displaying the grid
plt.grid(True)

# Show plot
plt.show()
```

```python
#histogram

plt.figure(figsize=(8, 6))  # Adjust figure size if necessary
plt.hist(growth_rate, bins=10, color='skyblue', edgecolor='black')  # Adjust number of bins as needed
plt.title('Distribution of Total Agriculture Growth Rate')
plt.xlabel('Growth Rate')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#scatter

plt.figure(figsize=(8, 6))  # Adjust figure size if necessary
plt.scatter(years, growth_rate, color='green', marker='o')
plt.title('Total Agriculture Growth Rate Over the Years')
plt.xlabel('Year')
plt.ylabel('Growth Rate')
plt.grid(True)
plt.show()
```

```python
#bar graph

plt.figure(figsize=(10, 6))  # Adjust figure size if necessary
plt.bar(years, growth_rate, color='orange')
plt.title('Total Agriculture Growth Rate Over the Years')
plt.xlabel('Year')
plt.ylabel('Growth Rate')
plt.grid(axis='y')
plt.show()
```
